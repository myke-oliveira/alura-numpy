### Baixando arquivos CSV do GitHub

Este código irá baixar os arquivos `apples_ts.csv`, `bytebank.csv` e `citrus.csv` do repositório GitHub fornecido e salvá-los no diretório atual do seu notebook Colab.

```python
import requests
import os

github_repo_url = "https://raw.githubusercontent.com/myke-oliveira/alura-numpy/main/"
csv_files = [
    "apples_ts.csv",
    "bytebank.csv",
    "citrus.csv"
]

# Criar um diretório para os arquivos CSV, se não existir
# csv_dir = "data"
# os.makedirs(csv_dir, exist_ok=True)

for filename in csv_files:
    file_url = github_repo_url + filename
    # file_path = os.path.join(csv_dir, filename) # Se quiser salvar em um subdiretório 'data'
    file_path = filename # Salvar no diretório atual

    if os.path.exists(file_path):
        print(f"'{filename}' já existe no diretório. Pulando o download.")
    else:
        print(f"Baixando {filename}...")
        response = requests.get(file_url)
        if response.status_code == 200:
            with open(file_path, 'wb') as f:
                f.write(response.content)
            print(f"'{filename}' baixado com sucesso!")
        else:
            print(f"Erro ao baixar '{filename}'. Status code: {response.status_code}")

print("\nVerificando os arquivos baixados:")
for filename in csv_files:
    # file_path = os.path.join(csv_dir, filename) # Se salvou em um subdiretório 'data'
    file_path = filename # Se salvou no diretório atual
    if os.path.exists(file_path):
        print(f"- {filename} existe (tamanho: {os.path.getsize(file_path)} bytes)")
    else:
        print(f"- {filename} NÃO foi encontrado.")
```

```python
import sys

# Let's check the size of the 'Moscow' array
size_moscow = sys.getsizeof(Moscow)
print(f"Size of 'Moscow' array: {size_moscow} bytes")

# Let's check the size of a simple Python list for comparison
my_list = [1, 2, 3, 4, 5]
size_list = sys.getsizeof(my_list)
print(f"Size of 'my_list': {size_list} bytes")
```

The `sys.getsizeof()` function returns the size of an object in bytes. Keep in mind that for complex objects like lists or custom classes, `getsizeof()` only accounts for the object's own memory consumption, not the memory used by objects it refers to. For NumPy arrays, it generally gives a good indication of the actual data size.

```python
import sys
import psutil
import os

# Get the current process
process = psutil.Process(os.getpid())

# Get memory info in bytes
mem_info = process.memory_info()

print(f"Uso de memória do script:")
print(f"  RSS (Resident Set Size): {mem_info.rss / (1024 * 1024):.2f} MB")
print(f"  VMS (Virtual Memory Size): {mem_info.vms / (1024 * 1024):.2f} MB")
```

O código foi executado com sucesso e retornou as seguintes informações de memória para o seu script:

*   **RSS (Resident Set Size)** - Esta é a porção de memória principal (RAM) que o processo está utilizando atualmente e que está armazenada na memória física. É a memória que não está trocada (swapped out) para o disco.
*   **VMS (Virtual Memory Size)** - Esta é a quantidade total de memória virtual que o processo reservou. Inclui toda a memória que o processo pode acessar, incluindo a memória RAM, memória de swap, e quaisquer bibliotecas compartilhadas carregadas. Geralmente, VMS é maior que RSS.

Esses valores fornecem uma visão geral do uso de memória do seu script Python no ambiente Colab.
