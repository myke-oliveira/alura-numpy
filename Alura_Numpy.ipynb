### Baixando arquivos CSV do GitHub

Este código irá baixar os arquivos `apples_ts.csv`, `bytebank.csv` e `citrus.csv` do repositório GitHub fornecido e salvá-los no diretório atual do seu notebook Colab.

```python
import requests
import os

github_repo_url = "https://raw.githubusercontent.com/myke-oliveira/alura-numpy/main/"
csv_files = [
    "apples_ts.csv",
    "bytebank.csv",
    "citrus.csv"
]

# Criar um diretório para os arquivos CSV, se não existir
# csv_dir = "data"
# os.makedirs(csv_dir, exist_ok=True)

for filename in csv_files:
    file_url = github_repo_url + filename
    # file_path = os.path.join(csv_dir, filename) # Se quiser salvar em um subdiretório 'data'
    file_path = filename # Salvar no diretório atual

    if os.path.exists(file_path):
        print(f"'{filename}' já existe no diretório. Pulando o download.")
    else:
        print(f"Baixando {filename}...")
        response = requests.get(file_url)
        if response.status_code == 200:
            with open(file_path, 'wb') as f:
                f.write(response.content)
            print(f"'{filename}' baixado com sucesso!")
        else:
            print(f"Erro ao baixar '{filename}'. Status code: {response.status_code}")

print("\nVerificando os arquivos baixados:")
for filename in csv_files:
    # file_path = os.path.join(csv_dir, filename) # Se salvou em um subdiretório 'data'
    file_path = filename # Se salvou no diretório atual
    if os.path.exists(file_path):
        print(f"- {filename} existe (tamanho: {os.path.getsize(file_path)} bytes)")
    else:
        print(f"- {filename} NÃO foi encontrado.")
```
